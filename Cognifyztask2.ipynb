```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
```

```python
restaurant_data=pd.read_csv(r"Dataset .csv")
restaurant_data
```

```python
restaurant_data.info()
restaurant_data.describe()
```

```python
restaurant_data.isnull().sum()
```

```python
restaurant_data['Cuisines'].unique()
restaurant_data['Price range'].unique()
restaurant_data['City'].unique()
```

```python
restaurant_data['Cuisines'].unique()
```

```python
label_encoders = {}
for i in ['Cuisines', 'Price range', 'City']:
    le = LabelEncoder()
    restaurant_data[i] = le.fit_transform(restaurant_data[i])
    label_encoders[i] = le
```

```python
#user prefernce
user_preferences = {}
user_preferences['Cuisines'] = input("Enter your favourite cuisine: ")
user_preferences['Price range'] = input("Enter your price range: ")
user_preferences['City'] = input("Enter your city: ")
```

```python
encoded_preferences = []
for column in ['Cuisines', 'Price range', 'City']:
    encoded_value = label_encoders[column].transform([user_preferences[column]])[0]
    encoded_preferences.append(encoded_value)

encoded_preferences
```

```python
encoded_preferences = np.array(encoded_preferences).reshape(1, -1)
encoded_preferences
```

```python
restaurant_features = restaurant_data[['Cuisines', 'Price range', 'City']].values
similarities = cosine_similarity(encoded_preferences, restaurant_features)
num_recommendations = 5
recommended_indices = similarities.argsort()[0][-num_recommendations:][::-1]
```

```python
recommended_restaurants = restaurant_data.iloc[recommended_indices]['Restaurant Name']
print("Recommended restaurants:")
print(recommended_restaurants)
```
