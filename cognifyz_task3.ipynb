```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
restaurant_data=pd.read_csv(r"Dataset .csv")
restaurant_data
```

```python
restaurant_data.info()
```

```python
columns_to_keep = ['Cuisines', 'Restaurant Name','Average Cost for two']
restaurant_data.drop(restaurant_data.columns.difference(columns_to_keep), axis=1,inplace=True)
```

```python
restaurant_data
```

```python
restaurant_data.isnull().sum()
```

```python
restaurant_data.dropna(inplace=True)
```

```python
restaurant_data.isnull().sum()
```

```python
missing_values = restaurant_data.isna().sum()
missing_values_column = restaurant_data['Restaurant Name'].isna().sum()
missing_values_column = restaurant_data['Cuisines'].isna().sum()
missing_values_column = restaurant_data['Average Cost for two'].isna().sum()
```

```python
label_encoder = LabelEncoder()
restaurant_data['Restaurant Name'] = label_encoder.fit_transform(restaurant_data['Restaurant Name'])
restaurant_data['Cuisines'] = label_encoder.fit_transform(restaurant_data['Cuisines'])
```

```python
x = restaurant_data[['Restaurant Name', 'Average Cost for two']]
y = restaurant_data['Cuisines']
```

```python
Scalar = StandardScaler()

Scalar.fit(x)
x = Scalar.transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
model_DT = DecisionTreeClassifier(random_state=10,criterion="gini")
model_DT.fit(x_train,y_train)
y_pred = model_DT.predict(x_test)
print(list(zip(y_test,y_pred)))
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
accuracy_score=accuracy_score(y_test,y_pred)
print("Accuracy of the model: ",accuracy_score)
```
